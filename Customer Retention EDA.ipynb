```python
x=data.drop(columns='3 Which city do you shop online from?',axis=1)
```

```python
## For Checking the Skewness
plt.figure(figsize=(50,80))
plotnumber=1
for column in x:
    if plotnumber<=40:
        ax=plt.subplot(10,4,plotnumber)
        sns.distplot(x[column])
        plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.show()
```

```python
plt.figure(figsize=(30,50))
graph=1
for column in x:
    if graph<=40:
        ax=plt.subplot(10,4,graph)
        sns.boxplot(x[column],orient='v')
        plt.xlabel(column,fontsize=10)
    graph+=1
plt.show()
```
